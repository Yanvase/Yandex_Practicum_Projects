## Описание проекта:
Вы работаете в интернет-магазине «Стримчик», который продаёт по всему миру компьютерные игры. Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). Вам нужно выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.
    
Перед вами данные до 2016 года. Представим, что сейчас декабрь 2016 г., и вы планируете кампанию на 2017-й. Нужно отработать принцип работы с данными. Неважно, прогнозируете ли вы продажи на 2017 год по данным 2016-го или же 2027-й — по данным 2026 года.
    
В наборе данных попадается аббревиатура ESRB (Entertainment Software Rating Board) — это ассоциация, определяющая возрастной рейтинг компьютерных игр. ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».

    

    
## Описание данных:
* Name — название игры
* Platform — платформа
* Year_of_Release — год выпуска
* Genre — жанр игры
* NA_sales — продажи в Северной Америке (миллионы проданных копий)
* EU_sales — продажи в Европе (миллионы проданных копий)
* JP_sales — продажи в Японии (миллионы проданных копий)
* Other_sales — продажи в других странах (миллионы проданных копий)
* Critic_Score — оценка критиков (максимум 100)
* User_Score — оценка пользователей (максимум 10)
* Rating — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.
    
    * «EC» («Early childhood») — «Для детей младшего возраста»
    * «E» («Everyone») — «Для всех»
    * «E10+» («Everyone 10 and older») — «Для всех старше 10 лет»
    * «T» («Teen») — «Подросткам 13—19 лет» 
    * «M» («Mature») — «От 17 лет»
    * «AO» («Adults Only 18+») — «Только для взрослых»
    * «RP» ( «Rating Pending»») — «Рейтинг ожидается»
    * «EC» («Early childhood»)
Данные за 2016 год могут быть неполными.

## Преобразование данных в нужные типы. Обработка пропусков.

```python
df_games['year_of_release'].isna().sum()
```

```python
# Обработаем 'year_of_release'

display(df_games['year_of_release'].describe())
plt.hist(df_games['year_of_release'], label='year_of_release', bins=100, alpha=0.5, range = (1980, 2016))
plt.suptitle("Пропуски")
plt.ylabel('Количество')
plt.xlabel('Год')
print('пропусков:', df_games['year_of_release'].isna().sum())

#заполним пропуски маркерным значением:
df_games['year_of_release'] = df_games['year_of_release'].fillna(1950).astype(int)
```

<b> Вывод: <b/>
    Хоть и встречаются пропуски, их доля не велика. Причина их возникновения не понятна. Возможно ошибка при выгрузке данных. Заполнять значения возможно маркером. Используем 1950.

```python
df_games['critic_score'].describe()
```

```python
# Обработаем 'critic_score'

display(df_games['critic_score'].describe())
plt.hist(df_games['critic_score'], label='critic_score', bins=100, alpha=0.5, range = (0, 100.))
plt.suptitle("Пропуски")
plt.ylabel('Количество')
plt.xlabel('Оценка критиков')
print('пропусков:', df_games['critic_score'].isna().sum())

#заполним пропуски маркерным значением:
df_games['critic_score'] = df_games['critic_score'].fillna(-1).astype(float)
```

<b> Вывод: <b/>
Заполнение медианным значением не приминимо так как оценка критиков может быть очень разная от игры к игре и повлиять на результат исследования. Заменим пропуски на маркер -1

```python

# в данном столбце присутсвует значение tbd, заменим на Nan:
df_games['user_score'] = df_games['user_score'].replace('tbd','NaN')

#поменяем тип данных на float:
df_games['user_score'] = df_games['user_score'].astype(float)

# Обработаем 'user_score'
display(df_games['user_score'].describe())
plt.hist(df_games['user_score'], label='user_score', bins=100, alpha=0.5, range = (0, 10.))
plt.suptitle("Пропуски")
plt.ylabel('Количество')
plt.xlabel('Оценка пользователей')
print('пропусков:', df_games['user_score'].isna().sum())

#заполним пропуски маркерным значением:
df_games['user_score'] = df_games['user_score'].fillna(-1)


```

<b> Вывод: <b/>
Как и в предыдущем, тут много пропусков, но заполнять их медианным значением не правильно. Заполним маркером -1

```python
# Обработаем 'Rating'
print('пропусков:', df_games['rating'].isna().sum())

#заполним пропуски маркерным значением:
df_games['rating'] = df_games['rating'].fillna('NOT_RATED')
```

<b> Вывод: <b/>
Заполним маркером 'NOT_RATED'

```python
# Обработаем 'name'
#Тут всего 2 игры и пропуски совпадают с пропусками в жанре. Лучше удалим.
df_games = df_games.dropna(subset = ['name'])
```

```python
df_games.info()
```

<b> Вывод: <b/>
После анализа была произведена замена на маркерные значения. Столбец хранящий рейтинг (rating) не будем менять так как нет возможности дополнить, а пропуски в данном столбце не повлияют на исследнование.

# Проверка гипотез:

<b> Для проверки использую специальный метод для проверки гипотезы (t-распределением Стьюдента) о равенстве среднего двух генеральных совокупностей по взятым из них выборкам так как эти генеральные совокупности не зависят друг от друга и их средние нормально распределены.

<a id='20-bullet'></a>
## Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;


H0 — Средние пользовательские рейтинги платформ Xbox One и PC одинаковые.

H1 — Средние пользовательские рейтинги платформ Xbox One и PC различаются.

alpha = 5%

```python
xbox_one = df_games.query('platform == "XOne" & user_score > 0 & year_of_release >= 2015')['user_score']
PC = df_games.query('"PC" in platform & user_score > 0 & year_of_release >= 2015')['user_score']

alpha = .05
results = st.ttest_ind(
    xbox_one, 
    PC)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

<a id='21-bullet'></a>
## Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.;

H0 — Средние пользовательские рейтинги жанров Action и Sports одинаковые.

H1 — Средние пользовательские рейтинги жанров Action и Sports различаются.

alpha = 5%

```python
action = df_games.query('genre == "Action" & user_score > 0 & year_of_release >= 2015')['user_score']
sports = df_games.query('genre == "Sports" & user_score > 0 & year_of_release >= 2015')['user_score']

alpha = .05
results = st.ttest_ind(
    action, 
    sports)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

<b> Вывод: <b/>
    
* Средние пользовательские рейтинги платформ Xbox One и PC одинаковые.
* Средние пользовательские рейтинги жанров Action и Sports различаются.

<a id='22-bullet'></a>
## Пояснение:

Гипотезы формировал по принципу H0 - старая гипотеза(равны), H1 - новая(отличаются). Для проверки использовал специальный метод для проверки гипотезы (t-распределением Стьюдента) о равенстве среднего двух генеральных совокупностей по взятым из них выборкам так как эти генеральные совокупности не зависят друг от друга и их средние нормально распределены.
